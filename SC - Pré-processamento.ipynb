```python
import pandas as pd
from datetime import date
```

```python
df = pd.read_csv('raw_dataset/covid_sc_20201207_2.csv', sep=';', low_memory=False)
```

```python
df
```

Faixa Etária

```python
for idx, row in df.iterrows():
    if  df.loc[idx,'idade'] < 1:
        df.loc[idx,'idade'] = 'Bebê'
    elif df.loc[idx,'idade'] < 9:
        df.loc[idx,'idade'] = 'Criança'
    elif df.loc[idx,'idade'] < 19:
        df.loc[idx,'idade'] = 'Jovem'
    elif df.loc[idx,'idade'] < 39:
        df.loc[idx,'idade'] = 'Adulto_1'
    elif df.loc[idx,'idade'] < 49:
        df.loc[idx,'idade'] = 'Adulto_2'
    elif df.loc[idx,'idade'] < 59:
        df.loc[idx,'idade'] = 'Adulto_3'
    elif df.loc[idx,'idade'] < 69:
        df.loc[idx,'idade'] = 'Idoso_1'
    elif df.loc[idx,'idade'] < 79:
        df.loc[idx,'idade'] = 'Idoso_2'
    else:
        df.loc[idx,'idade'] = 'Idoso_3'
```

CRITERIO

```python
for idx, row in df.iterrows():
    if  df.loc[idx,'tipo_teste'] == 'BIOLOGIA MOLECULAR (RT-PCR)':
        df.loc[idx,'tipo_teste'] = 'RT-PCR'
    elif df.loc[idx,'tipo_teste'] == 'IMUNOLOGICO (TESTE RAPIDO)':
        df.loc[idx,'tipo_teste'] = 'TESTE RÁPIDO'
    else:
        df.loc[idx,'tipo_teste'] = 'Outros'
```

ETNIA

```python
for idx, row in df.iterrows():
    if  df.loc[idx,'raca'] == 'IGNORADO':
        df.loc[idx,'raca'] = 'NAO INFORMADO'
```

SINTOMAS

```python
df['FEBRE'] = df.apply(lambda row: 'SIM' if 'FEBRE' in row.sintomas else 'NAO', axis=1)
df['TOSSE'] = df.apply(lambda row: 'SIM' if 'TOSSE' in row.sintomas else 'NAO', axis=1)
df['GARGANTA'] = df.apply(lambda row: 'SIM' if 'DOR DE GARGANTA' in row.sintomas else 'NAO', axis=1)
df['DISPNEIA'] = df.apply(lambda row: 'SIM' if 'DISPNEIA' in row.sintomas else 'NAO', axis=1)
df['Outros'] = df.apply(lambda row: 'SIM' if 'DIARREIA' or 'DOR NO CORPO' or 'CORIZA' or 'MIALGIA' or 'CANSACO' or 'CEFALEIA' or 'CONGESTAO NASAL' in row.sintomas else 'NAO', axis=1)
del df['sintomas']
```

```python
df['asma'] = df.apply(lambda row: 'SIM' if 'ASMA' in row.comorbidades else 'NAO', axis=1)
df['d_respiratoria'] = df.apply(lambda row: 'SIM' if 'PNEUMATICA' in row.comorbidades else 'NAO', axis=1)
df['outras_comorbidades'] = df.apply(lambda row: 'SIM' if 'CANCER' or 'HIPERTENSAO' in row.comorbidades else 'NAO', axis=1)
df['d_neurologica'] = df.apply(lambda row: 'SIM' if 'NEUROLOGICA' in row.comorbidades else 'NAO', axis=1)
df['diabete'] = df.apply(lambda row: 'SIM' if 'DIABETES' in row.comorbidades else 'NAO', axis=1)
df['d_neurologica'] = df.apply(lambda row: 'SIM' if 'NEUROLOGICA' in row.comorbidades else 'NAO', axis=1)
df['imunodeficiencia'] = df.apply(lambda row: 'SIM' if 'IMUNODEPRESSAO' in row.comorbidades else 'NAO', axis=1)
df['d_renal'] = df.apply(lambda row: 'SIM' if 'RENAL' in row.comorbidades else 'NAO', axis=1)
df['d_hepatica'] = df.apply(lambda row: 'SIM' if 'HEPATICA' in row.comorbidades else 'NAO', axis=1)
df['d_hematologica'] = df.apply(lambda row: 'SIM' if 'HEMATOLOGICA' in row.comorbidades else 'NAO', axis=1)
df['d_cardiaca'] = df.apply(lambda row: 'SIM' if 'CARDIOVASCULAR' in row.comorbidades else 'NAO', axis=1)
df['obesidade'] = df.apply(lambda row: 'SIM' if 'OBESIDADE' in row.comorbidades else 'NAO', axis=1)
df['sindrome_down'] = df.apply(lambda row: 'SIM' if 'SINDROME DE DOWN' in row.comorbidades else 'NAO', axis=1)
del df['comorbidades']
```

INTERNAÇÃO

```python
dt_conf = pd.to_datetime(df['data_internacao'])
dt_obito = pd.to_datetime(df['data_obito'])
df['INTERNACAO'] = round(((dt_obito - dt_conf).dt.days),-1)
# del df['dt_conf']
# del df['dt_obito']
```

```python
for idx, row in df.iterrows():
    if pd.isna(df.loc[idx,'INTERNACAO']):
        df.loc[idx,'INTERNACAO'] = 'NAO DISPONIVEL'
```

```python
for index, row in df.iterrows():
    mes = row['data_resultado'].split('-')
    if mes[1] == '02':
        row['data_resultado'] = 'Fevereiro'
    elif mes[1] == '03':
        row['data_resultado'] = 'Março'
    elif mes[1] == '04':
        row['data_resultado'] = 'Abril'
    elif mes[1] == '05':
        row['data_resultado'] = 'Maio'    
    elif mes[1] == '06':
        row['data_resultado'] = 'Junho'
    elif mes[1] == '07':
        row['data_resultado'] = 'Julho'
    elif mes[1] == '08':
        row['data_resultado'] = 'Agosto'
    elif mes[1] == '09':
        row['data_resultado'] = 'Setembro'
    elif mes[1] == '10':
        row['data_resultado'] = 'Outubro'
    elif mes[1] == '11':
        row['data_resultado'] = 'Novembro'
    else:
        row['data_resultado'] = 'Dezembro'
```

```python
for index, row in df.iterrows():
    mes = row['data_obito'].split('-')
    if mes[1] == '02':
        row['data_obito'] = 'Fevereiro'
    elif mes[1] == '03':
        row['data_obito'] = 'Março'
    elif mes[1] == '04':
        row['data_obito'] = 'Abril'
    elif mes[1] == '05':
        row['data_obito'] = 'Maio'    
    elif mes[1] == '06':
        row['data_obito'] = 'Junho'
    elif mes[1] == '07':
        row['data_obito'] = 'Julho'
    elif mes[1] == '08':
        row['data_obito'] = 'Agosto'
    elif mes[1] == '09':
        row['data_obito'] = 'Setembro'
    elif mes[1] == '10':
        row['data_obito'] = 'Outubro'
    elif mes[1] == '11':
        row['data_obito'] = 'Novembro'
    else:
        row['data_obito'] = 'Dezembro'
```

```python
for idx, row in df.iterrows():
    if  df.loc[idx,'internacao'] == 'INTERNADO':
        df.loc[idx,'internacao'] = 'SIM'
    else:
        df.loc[idx,'internacao'] = 'NAO'

for idx, row in df.iterrows():
    if  df.loc[idx,'internacao_uti'] == 'INTERNADO UTI':
        df.loc[idx,'internacao_uti'] = 'SIM'
    else:
        df.loc[idx,'internacao_uti'] = 'NAO'
```

```python
for idx, row in df.iterrows():
    if pd.isna(df.loc[idx,'gestante']):
        df.loc[idx,'gestante'] = 'NAO'
    else:
        df.loc[idx,'gestante'] = 'SIM'
```
