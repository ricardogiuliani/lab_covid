```python
import pandas as pd
import import_ipynb
import covid_dict as cd
```

```python
dic = cd.covid_dict
```

```python
df = pd.read_csv('input_datasets/no_death[1].csv', header=None, sep=',')
column_sum = df.sum(axis=0)

zeros = []
for i in range(0, len(column_sum)):
    if column_sum[i] == 0:
        zeros.append(i)

removed = []
for each in zeros: 
     removed.append(dic[each])

for e in zeros:
    del df[e]
    
for col,row in df.iteritems():
    df.loc[df[col] == 0, col] = 0
    df.loc[df[col] == 1, col] = int(df.columns.get_loc(col))

panda = df.to_numpy()

res = [[ele for ele in sub if ele != 0] for sub in panda]

pd.DataFrame(res).to_csv("input_datasets/panda/no_death[2].csv", header=None, index=False, sep =" ")
```
