```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('20201030_data_visualisation.csv')
```

```python
def roundup(x):
    return 100 + int(math.ceil(x / 100.0)) * 100 
```

```python
def graph(attr, rotation = False, h = 4):
    sns.catplot(x=attr, data=df, kind="count", height=h, aspect=1)
    ax = plt.gca()
    
    if (rotation):
        plt.xticks(rotation=90)
        
    y_max = df[attr].value_counts().max() 
    ax.set_ylim([0, roundup(y_max)])

    for p in ax.patches:
        ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
                fontsize=12, color='black', ha='center', va='bottom')
```

```python
graph('REGIAO_COVID', rotation=True, h=8)
graph('SEXO')
graph('FAIXAETARIA', rotation=True, h=6)
```
