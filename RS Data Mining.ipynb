```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
hosp = pd.read_csv("raw_dataset/covid_dataset/Hospitalizados.csv", sep=';')
bed = pd.read_csv("raw_dataset/covid_dataset/Leitos.csv", sep=';')
rec = pd.read_csv("raw_dataset/covid_dataset/Recuperados.csv", sep=';')
iso = pd.read_csv("raw_dataset/covid_dataset/Isolamento.csv", sep=';')
```

### Hospitalizados

```python
hosp = hosp[hosp['fu_code'] == 43]
hosp.rename(columns={"type_of_hospitalization": "Tipo de Hospitalização"}, inplace=True)

for index in hosp.index:
    if hosp.loc[index, 'Tipo de Hospitalização'] == 1:
        hosp.loc[index, 'Tipo de Hospitalização'] = 'UTI'
    elif hosp.loc[index, 'Tipo de Hospitalização'] == 2:
        hosp.loc[index, 'Tipo de Hospitalização'] = 'Clínico'
```

```python
plt.figure(figsize=(30, 10))
sns.lineplot(x="date", y="confirmed", hue='Tipo de Hospitalização', data=hosp, palette=['red','blue'])
```

### Leitos

```python
bed = bed[bed['fu_code'] == 43]
bed.rename(columns={"bed_type": "Tipo de Leito"}, inplace=True)

for index in bed.index:
    if bed.loc[index, 'Tipo de Leito'] == 1:
        bed.loc[index, 'Tipo de Leito'] = 'UTI'
    elif bed.loc[index, 'Tipo de Leito'] == 2:
        bed.loc[index, 'Tipo de Leito'] = 'Clínico'
```

```python
plt.figure(figsize=(30, 10))
sns.lineplot(x="date", y="quantity", hue='Tipo de Leito',
             data=bed,palette=['red','blue'])
```

### Recuperados

```python
rec = rec[rec['fu_code'] == 43]
```

```python
plt.figure(figsize=(30, 10))
sns.lineplot(x="date", y="Number",
             data=rec)
```

### Isolamento

```python
iso = iso[iso['cod_uf'] == 43]
```

```python
plt.figure(figsize=(30, 10))
sns.lineplot(x="date(yyyy-mm-dd)", y="indice",
             data=iso)
```

Data Visualisation

```python
df = pd.read_csv('20201030_data_visualisation.csv')
```

```python
def roundup(x):
    return 100 + int(math.ceil(x / 100.0)) * 100 
```

```python
def graph(attr, rotation = False, h = 4):
    sns.catplot(x=attr, data=df, kind="count", height=h, aspect=1)
    ax = plt.gca()
    
    if (rotation):
        plt.xticks(rotation=90)
        
    y_max = df[attr].value_counts().max() 
    ax.set_ylim([0, roundup(y_max)])

    for p in ax.patches:
        ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
                fontsize=12, color='black', ha='center', va='bottom')
```
