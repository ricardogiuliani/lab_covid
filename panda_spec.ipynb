```python
df = pd.read_csv('panda.csv', header=None, sep=' ')
df.head()
```

```python
death = df[df.eq(538).any(axis=1)]
```

```python
panda = death.to_numpy()
```

```python
pd.DataFrame(panda).to_csv("panda_obito.csv", header=None, index=False, sep =" ")
```
